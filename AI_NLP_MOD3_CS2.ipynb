```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tag import pos_tag
from nltk.chunk import ne_chunk
from nltk.tree import Tree
import matplotlib.pyplot as plt
import pandas as pd
```

ABC Company wants to perform some text analysis and make visualization for one of
their dataset.
The dataset has been taken from Kaggle.
(https://www.kaggle.com/crowdflower/twitter-airline-sentiment/home)
This is a dataset having tweets about 6 US Airlines along with their sentiments:
positive, negative and neutral.
You are provided with this dataset named “Tweets.csv”.
This is a snapshot of the dataset:
It has tweets in ‘text’ column and sentiments in ‘airline_sentiment’ column.


```python
#1. Retrieve all tags starting from ‘@’ in the entire dataset and save in a file called “References.txt”
def RetrieveTagsStartingWithAt(text):
    # Split the text into words
    words =text
    print("Words in the text:", words)
    
    # Filter words that start with '@']
    tags = []
    for word in words.str.split():
        for w in word:
            if w[0].startswith('@'):
                tags.append(w)
                break
    
    with open("References.txt", "w",encoding="utf-8") as file:
        for tag in set(tags):
            file.write(tag + "\n")

    return tags
```

```python
# 2. Extract all noun phrases from their dataset and save them in different lines in a
# file named “Noun Phrases for <airline_sentiment> Review .txt” (You can choose
# your own grammar for noun phrase).
# Here <airline_sentiment> will have 3 different values: positive, negative and
# neutral so 3 files will be created.
def ExtractNounPhrases(df):
    for sentiment in df['airline_sentiment'].unique():
        noun_phrases = []
        for text in df[df['airline_sentiment'] == sentiment]['text']:
            words = word_tokenize(text)
            tagged_words = pos_tag(words)
            grammar = 'NP: {<DT>?<JJ>*<NN.*>+}'
            cp = ne_chunk(tagged_words, binary=False)
            for subtree in cp.subtrees():
                if subtree.label() == 'NP':
                    noun_phrases.append(' '.join(word for word, tag in subtree.leaves()))
        
        with open(f"Noun Phrases for {sentiment} Review.txt", "w", encoding="utf-8") as file:
            for np in set(noun_phrases):
                file.write(np + "\n")
```

```python
# 3. Extract all verb phrases from their dataset and save them in different lines in a file
# named “Verb Phrases for <airline_sentiment> Review .txt”” (You can choose your
# own grammar for noun phrase).
# Here <airline_sentiment> will have 3 different values: positive , negative and
# neutral so 3 files will be created.
def ExtractVerbPhrases(df):
    for sentiment in df['airline_sentiment'].unique():
        verb_phrases = []
        for text in df[df['airline_sentiment'] == sentiment]['text']:
            words = word_tokenize(text)
            tagged_words = pos_tag(words)
            grammar = 'VP: {<VB.*><NP|PP|CLAUSE>+$}'
            cp = ne_chunk(tagged_words, binary=False)
            for subtree in cp.subtrees():
                if subtree.label() == 'VP':
                    verb_phrases.append(' '.join(word for word, tag in subtree.leaves()))
        
        with open(f"Verb Phrases for {sentiment} Review.txt", "w", encoding="utf-8") as file:
            for vp in set(verb_phrases):
                file.write(vp + "\n")
```

```python
# 4. For each sentiment, make a well labelled pie chart showing distribution of Noun
# Phrases and Verb Phrases of that sentiment from the data set. Use the above
# created files to get the frequencies.
def CreatePieCharts(df):
    for sentiment in df['airline_sentiment'].unique():
        with open(f"Noun Phrases for {sentiment} Review.txt", "r", encoding="utf-8") as file:
            noun_phrases = file.readlines()
        
        with open(f"Verb Phrases for {sentiment} Review.txt", "r", encoding="utf-8") as file:
            verb_phrases = file.readlines()
        
        labels = ['Noun Phrases', 'Verb Phrases']
        sizes = [len(noun_phrases), len(verb_phrases)]
        
        plt.figure(figsize=(6, 6))
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
        plt.title(f'Distribution of Noun and Verb Phrases for {sentiment} Reviews')
        plt.axis('equal')
        plt.show()
```

```python
#=======================================Q1=======================================
csvFile=pd.read_csv("Tweets.csv")
df=pd.DataFrame(csvFile[columns:=['text', 'airline_sentiment']])
# 1. Retrieve all tags starting from ‘@’ in the entire dataset and save in a file called “References.txt”
tags=RetrieveTagsStartingWithAt(df['text'])
print("Tags starting with '@' have been saved to 'References.txt'.", set(tags))
#=======================================Q2=======================================
ExtractNounPhrases(df)
print("Noun phrases have been extracted and saved to respective files based on sentiment.")
#=======================================Q3=======================================
ExtractVerbPhrases(df)
print("Verb phrases have been extracted and saved to respective files based on sentiment.")
#=======================================Q4=======================================
CreatePieCharts(df)
print("Pie charts for distribution of noun and verb phrases have been created for each sentiment.")
```
